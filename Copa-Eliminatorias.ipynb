```python
import random
import copy
from datetime import datetime

class Team:
  def __init__(self, name, wins, games):
    self._name = name
    self._wins = wins
    self._games = games
    if(wins != 0 and games != 0):
        self._win_rate = wins/games
    else:
        self._win_rate = 0
        
  def update_win_rate(self):
    self._win_rate = self._wins/self._games    
    
  def win(self):   
    self._games = self._games + 1
    self._wins = self._wins + 1
    self.update_win_rate()  
    
  def lose(self):
    self._games = self._games + 1 
    self.update_win_rate() 
    
  def tie(self):
    self._games = self._games + 1 
    self.update_win_rate()       
    
  def __repr__(self):
    return (self._name +" : "+ str(self._wins)+ " / "+ str(self._games)+ " / "+ str(self._win_rate) + "\n")

def simulate_game_elm(team_one,team_two,match_history,brazil_history,current_stage,brazil_encounter):
  status = "UNDEFINED"
  while(status == "UNDEFINED" or status == "TIE"):
        rng = random.uniform(0,1)
        rng_two = random.uniform(0,1)  
        
        if(team_one._win_rate >= rng):
            if(team_two._win_rate < rng_two):      
              team_one.win()  
              status = "WIN"
              if(team_one._name  == "Brasil" or team_two._name == "Brasil"):
                match_history.append(team_one._name+" X "+ team_two._name)                             
                found_brazil_history = False
                if(team_one._name == "Brasil"):
                    current_match = team_one._name+" | "+team_two._name + ": WIN"
                    current_encounter = team_one._name+" X "+team_two._name     
                else:
                    current_match = team_two._name+" | "+team_one._name + ": LOSE"
                    current_encounter = team_two._name+" X "+team_one._name    
                for matchAndCounter in brazil_history:
                    if(matchAndCounter[0] == current_match):
                        found_brazil_history = True
                        matchAndCounter[1] = matchAndCounter[1] + 1
                if(not found_brazil_history):
                    brazil_history.append([current_match,1])     
                found_brazil_encounter = False               
                for matchAndCounter in brazil_encounter:
                    if(matchAndCounter[0] == current_encounter):
                        found_brazil_encounter = True
                        matchAndCounter[1] = matchAndCounter[1] + 1
                if(not found_brazil_encounter):
                    brazil_encounter.append([current_encounter,1,current_stage])   
              return team_one
            else:     
              status = "TIE"
        else:
            if(team_two._win_rate >= rng_two):      
              team_two.win()              
              status = "WIN"
              if(team_one._name  == "Brasil" or team_two._name == "Brasil"):              
                match_history.append(team_one._name+" X "+ team_two._name)                   
                found_brazil_history = False
                if(team_one._name == "Brasil"):
                    current_encounter = team_one._name+" X "+team_two._name
                    current_match = team_one._name+" | "+team_two._name + ": LOSE"
                else:
                    current_encounter = team_two._name+" X "+team_one._name  
                    current_match = team_two._name+" | "+team_one._name + ": WIN"                
                for matchAndCounter in brazil_history:
                    if(matchAndCounter[0] == current_match):
                        found_brazil_history = True
                        matchAndCounter[1] = matchAndCounter[1] + 1
                if(not found_brazil_history):
                    brazil_history.append([current_match,1])
                found_brazil_encounter = False               
                for matchAndCounter in brazil_encounter:
                    if(matchAndCounter[0] == current_encounter):
                        found_brazil_encounter = True
                        matchAndCounter[1] = matchAndCounter[1] + 1
                if(not found_brazil_encounter):
                    brazil_encounter.append([current_encounter,1,current_stage])                     
              return team_two
            else:      
              status = "TIE"            

def initialize_list(original_team_list):
    current_file = "second_phase.txt"
    original_team_list = []   
    f = open(current_file, "r",encoding="utf-8-sig")        
    for line in f.readlines():
        ele = line.split(";")
        if(ele[0] != '\n'):
            ele[3] = ele[3][:-1]         
            temp_team = Team(ele[0],int(ele[2]),int(ele[3]))
            original_team_list.append(temp_team)
    return original_team_list
```

```python
%%time

k = 10000000
second_phase_teams = []
brazil_history = []
brazil_encounter = []
brazil_matches_frames = []
brazil_matches_sucess_frames = []
coup_winners = []
            
for i in range(0,k):    
  second_phase_teams = initialize_list(second_phase_teams)     
  
  brazil_matches = []  
  first_a_team = second_phase_teams[0]
  second_a_team = second_phase_teams[1]
  first_b_team = second_phase_teams[2]
  second_b_team = second_phase_teams[3]
  first_c_team = second_phase_teams[4]
  second_c_team = second_phase_teams[5]
  first_d_team = second_phase_teams[6]
  second_d_team = second_phase_teams[7]    
  first_e_team = second_phase_teams[8]
  second_e_team = second_phase_teams[9]
  first_f_team = second_phase_teams[10]
  second_f_team = second_phase_teams[11]
  first_g_team = second_phase_teams[12]
  second_g_team = second_phase_teams[13]
  first_h_team = second_phase_teams[14]
  second_h_team = second_phase_teams[15]  

  team_49 = simulate_game_elm(first_a_team,second_b_team,brazil_matches,brazil_history,"Oitava",brazil_encounter)     
  team_50 = simulate_game_elm(first_c_team,second_d_team,brazil_matches,brazil_history,"Oitava",brazil_encounter) 
  team_51 = simulate_game_elm(first_d_team,second_c_team,brazil_matches,brazil_history,"Oitava",brazil_encounter)   
  team_52 = simulate_game_elm(first_b_team,second_a_team,brazil_matches,brazil_history,"Oitava",brazil_encounter)    
  team_53 = simulate_game_elm(first_e_team,second_f_team,brazil_matches,brazil_history,"Oitava",brazil_encounter)  
  team_54 = simulate_game_elm(first_g_team,second_h_team,brazil_matches,brazil_history,"Oitava",brazil_encounter) 
  team_55 = simulate_game_elm(first_f_team,second_e_team,brazil_matches,brazil_history,"Oitava",brazil_encounter)   
  team_56 = simulate_game_elm(first_h_team,second_g_team,brazil_matches,brazil_history,"Oitava",brazil_encounter)   
    
  team_58 = simulate_game_elm(team_53,team_54,brazil_matches,brazil_history,"Quarta",brazil_encounter)  
  team_57 = simulate_game_elm(team_49,team_50,brazil_matches,brazil_history,"Quarta",brazil_encounter) 
  team_60 = simulate_game_elm(team_55,team_56,brazil_matches,brazil_history,"Quarta",brazil_encounter)   
  team_59 = simulate_game_elm(team_51,team_52,brazil_matches,brazil_history,"Quarta",brazil_encounter) 
    
  team_61 = simulate_game_elm(team_57,team_58,brazil_matches,brazil_history,"Semi",brazil_encounter)   
  team_62 = simulate_game_elm(team_59,team_60,brazil_matches,brazil_history,"Semi",brazil_encounter)
    
  team_63 = simulate_game_elm(team_61,team_62,brazil_matches,brazil_history,"Final",brazil_encounter) 

  found_matches = False
  current_match_hist = ""
  for match in brazil_matches:
    current_match_hist = current_match_hist + match + "; "  
  for histAndCounterAndCounter in brazil_matches_frames:     
     if(histAndCounterAndCounter[0] == current_match_hist):
         if(team_63._name == "Brasil"):
            found_matches = True
            histAndCounterAndCounter[1] = histAndCounterAndCounter[1] + 1
            histAndCounterAndCounter[2] = histAndCounterAndCounter[2] + 1
         else:
            found_matches = True
            histAndCounterAndCounter[1] = histAndCounterAndCounter[1] + 1
  if(not found_matches):
    if(team_63._name == "Brasil"):
        brazil_matches_frames.append([current_match_hist,1,1])
    else:
        brazil_matches_frames.append([current_match_hist,1,0])
     
  if(team_63._name == "Brasil"):
    found_matches_sucess = False
    current_match_sucess_hist = ""
    for match in brazil_matches:
        current_match_sucess_hist = current_match_sucess_hist + match + "; "  
    for histAndCounter in brazil_matches_sucess_frames:     
     if(histAndCounter[0] == current_match_sucess_hist):       
        found_matches_sucess = True
        histAndCounter[1] = histAndCounter[1] + 1
    if(not found_matches_sucess):    
        brazil_matches_sucess_frames.append([current_match_sucess_hist,1])

  found_winner = False
  for teamAndCounter in coup_winners:     
     if(teamAndCounter[0] == team_63._name):
         found_winner = True
         teamAndCounter[1] = teamAndCounter[1] + 1
  if(not found_winner):
     coup_winners.append([team_63._name,1])

brazil_encounter.sort(key=lambda x: x[1], reverse=True)     
for hist in brazil_encounter:
    hist[1] = (hist[1] / k) * 100 
    
phases = ["Oitava","Quarta","Semi","Final"]
for phase_match in phases:
    print(phase_match)
    current_total = 0
    for encounter in brazil_encounter:
        if(encounter[2] == phase_match):
            current_total = current_total + encounter[1]
    for encounter in brazil_encounter:
        if(encounter[2] == phase_match):
            encounter.append(100 * encounter[1] / current_total) 
        
brazil_history.sort(key=str,reverse=True)    
for hist in brazil_history:
    hist[1] = (hist[1] / k) * 100       
   
for i in range(len(brazil_history)):   
    if(i+1 < len(brazil_history)):
        if(brazil_history[i][0][:-4] == brazil_history[i+1][0][:-5]):
            brazil_history[i][1] = brazil_history[i][1] / (brazil_history[i][1] + brazil_history[i+1][1])
            del brazil_history[i+1]
        else:
            brazil_history[i][1] = 100
    
coup_winners.sort(key=lambda x: x[1], reverse=True)   
for group in coup_winners:
    group[1] = (group[1] / k) * 100
  
brazil_matches_sucess_frames.sort(key=lambda x: x[1], reverse=True)   
for group in brazil_matches_sucess_frames:
    group[1] = (group[1] / k) * 100   
    
brazil_matches_frames.sort(key=lambda x: x[2], reverse=True)   
for group in brazil_matches_frames:
    group[2] = (group[2] / group[1]) * 100
    group[1] = (group[1] / k) * 100     

dt = datetime.now()

file_string = "logs_elm/campeoes_" + str(dt)
with open(file_string, 'w', encoding="utf-8") as f:
    f.write("Campeões:\n")
    for frame in coup_winners:
        f.write("-> "+frame[0]+" "+str(frame[1])+"%.\n")  
f.close()

file_string = "logs_elm/caminho_brasileiro_sucesso_" + str(dt)
with open(file_string, 'w', encoding="utf-8") as f:        
    brazil_matches_frames.sort(key=lambda x: x[2], reverse=True)   
    f.write("Quadros e porcentagens do caminho Brasileiro: [ORDENADO POR SUCESSO]\n")
    for frame in brazil_matches_frames:
        if(frame[0] == ""):
            f.write("-> Não classificou. "+str(frame[1])+"%\n")
        else:
            f.write("-> "+frame[0]+" "+str(frame[1])+"%.\n\tSucesso: "+str(frame[2])+"%.\n")
f.close()

file_string = "logs_elm/caminho_brasileiro_ocorrencia_" + str(dt)
with open(file_string, 'w', encoding="utf-8") as f:  
    brazil_matches_frames.sort(key=lambda x: x[1], reverse=True)   
    f.write("Quadros e porcentagens do caminho Brasileiro: [ORDENADO POR OCORRÊNCIA]\n")
    for frame in brazil_matches_frames:
        if(frame[0] == ""):
            f.write("-> Não classificou. "+str(frame[1])+"%\n")
        else:
            f.write("-> "+frame[0]+" "+str(frame[1])+"%.\n\tSucesso: "+str(frame[2])+"%.\n")
f.close()

file_string = "logs_elm/caminho_brasileiro_vitorioso_" + str(dt)
with open(file_string, 'w', encoding="utf-8") as f:  
    f.write("Quadros e porcentagens do caminho vitorioso Brasileiro: [ORDENADO POR OCORRÊNCIA]\n")
    for frame in brazil_matches_sucess_frames:  
        f.write("-> "+frame[0]+" "+str(frame[1])+"%.\n")
f.close()

file_string = "logs_elm/historico_por_equipe_" + str(dt)
with open(file_string, 'w', encoding="utf-8") as f:  
    f.write("Histórico de desempenho contras as seleções:\n")
    for frame in brazil_history:  
        f.write("-> "+frame[0]+" "+str(frame[1])+"%.\n")
f.close()

file_string = "logs_elm/encontro_por_equipe_" + str(dt)
with open(file_string, 'w', encoding="utf-8") as f:  
    f.write("Histórico de encontra das seleções com o Brasil:\n")
    for frame in brazil_encounter:  
        f.write("-> "+frame[0]+". Ocorre (% geral) "+str(frame[1])+"%. Ocorre (% da "+str(frame[2])+") "+str(frame[3])+"%.\n")
f.close()

for item in brazil_encounter:
    print(item)
```
